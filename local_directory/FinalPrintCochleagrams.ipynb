# Visualizing the cochleagrams

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable

coch_example_single = np.load('./Room_01_RT60_1.0_gx_1.5_gy_3.5_gz_1.5_Az_000_El_045_natural_chirping_birds_8.npz')
coch_example_single = coch_example_single['arr_0'].astype(np.float32)
scoch_left = coch_example_single[0]
scoch_right = coch_example_single[1]

coch_example_multi = np.load('./Room_01_RT60_1.0_gx_1.5_gy_3.5_gz_1.5_Az_000_El_-20_Az_290_El_060_domestic_keyboard_typing_9_urban_helicopter_5.npz')
coch_example_multi = coch_example_multi['arr_0'].astype(np.float32)
mcoch_left = coch_example_multi[0]
mcoch_right = coch_example_multi[1]
```

```python
#Cochleagram of single-source data
fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=True)
im1 = ax1.imshow(scoch_left, aspect='auto', origin='lower', cmap='jet')
im2 = ax2.imshow(scoch_right, aspect='auto', origin='lower', cmap='jet')

divider = make_axes_locatable(ax1)
cax = divider.append_axes('right', size='2%', pad=0.15)
fig.colorbar(im1, cax=cax, orientation='vertical')

divider = make_axes_locatable(ax2)
cax = divider.append_axes('right', size='2%', pad=0.15)
fig.colorbar(im2, cax=cax, orientation='vertical')

# separate subplot titles
ax1.set_title('Cochleagram Left')
ax2.set_title('Cochleagram Right')

# common axis labels
fig.supxlabel('Time')
fig.supylabel('Frequency')
```

```python
#Cochleagram of single-source data
fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=True)
im1 = ax1.imshow(mcoch_left, aspect='auto', origin='lower', cmap='jet')
im2 = ax2.imshow(mcoch_right, aspect='auto', origin='lower', cmap='jet')

divider = make_axes_locatable(ax1)
cax = divider.append_axes('right', size='2%', pad=0.15)
fig.colorbar(im1, cax=cax, orientation='vertical')

divider = make_axes_locatable(ax2)
cax = divider.append_axes('right', size='2%', pad=0.15)
fig.colorbar(im2, cax=cax, orientation='vertical')

# separate subplot titles
ax1.set_title('Cochleagram Left')
ax2.set_title('Cochleagram Right')

# common axis labels
fig.supxlabel('Time')
fig.supylabel('Frequency')
```
